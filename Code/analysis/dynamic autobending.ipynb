### Dynamic Autobending dataset Period:  0.3 sec

After plate would bend 90 deg, 0.3 seconds wait before collecting 15 data points.  Same when going back to 0 degrees.  Test was conducted over 100 cycles

```python
DS1_datafiles = os.path.join(path_to_repository , "CSV Data/12_18_2024/dynamic 100 cycles/*.csv")

g = bender_class()
g.load_data(DS1_datafiles)
g.normalize_adc_over_R0()
g.dynamic_data(0.3) # time period after plate bending 90 deg or back to 0 deg before collecting 15 data points
g.plot_dynamic(12) # plot second subplot to 12 sec
```
