```python
import json
from pathlib import Path
import ast

p = Path(r"C:\Users\toppe\OneDrive - CSU Maritime Academy\Documents\GitHub\Strain-Sensor-\Code\analysis\camera_sensor_imu_analysis\adc_vs_cam_tom_2.ipynb")
txt = p.read_text(encoding="utf-8", errors="replace")

# Backup
p.with_suffix(".ipynb.bak").write_text(txt, encoding="utf-8")

# Quick corruption checks
if "<<<" in txt and ">>>" in txt:
    raise RuntimeError("Looks like a Git merge conflict is inside the notebook (<<<<<<< >>>>>>>). Resolve conflict first.")
if txt.lstrip().startswith("version https://git-lfs.github.com/spec/v1"):
    raise RuntimeError("This file is a Git LFS pointer, not the real notebook. Run: git lfs pull")

s = txt.strip()  # <-- key fix vs lstrip()

# If the whole file is a quoted string, unquote it
if (s.startswith("'") and s.endswith("'")) or (s.startswith('"') and s.endswith('"')):
    unquoted = ast.literal_eval(s)      # unescapes \r\n etc.
    nb = json.loads(unquoted)           # validates JSON
    p.write_text(json.dumps(nb, indent=1), encoding="utf-8")
    print("Repaired: unwrapped quoted JSON string.")
else:
    # If it already starts with {, validate directly
    if s.startswith("{"):
        nb = json.loads(txt)
        print("Notebook already valid JSON (maybe a different issue).")
    else:
        print("Not a single quoted JSON string. Open the file in a text editor and check the first line for: quotes, conflict markers, LFS pointer, or HTML.")
```
