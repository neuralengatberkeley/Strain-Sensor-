```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

calib_df = out_cam.get("calib_df", None)
coeffs   = out_cam.get("coeffs", None)

# ---- 1) Skip unless ADAB pairing ----
allowed = {"B1_adab", "B2_adab"}
if {folder_suffix_first, folder_suffix_second} != allowed:
    print("SKIP: suffixes are not exactly {'B1_adab','B2_adab'}")
else:

    # ---- 2) Ensure we have self-calibration coeffs for set 1 and set 2 ----
    if not isinstance(coeffs, dict):
        coeffs = {}

    if (1 not in coeffs or 2 not in coeffs):
        if calib_df is None or getattr(calib_df, "empty", True):
            print("WARNING: calib_df is empty/None and coeffs missing for set 1/2 -> cannot compute angles.")
        else:
            poly_order_self = 2
            for s in (1, 2):
                sub = calib_df[(calib_df["set"] == s) & calib_df["angle_snap_deg"].notna()]
                if len(sub) >= (poly_order_self + 1):
                    x_fit = sub["adc_mean"].to_numpy(dtype=float)
                    y_fit = sub["angle_snap_deg"].to_numpy(dtype=float)
                    coeffs[s] = np.polyfit(x_fit, y_fit, poly_order_self)

    # ---- 3) Build cross-trained coeffs: set 1 shape -> normalized -> rescaled to set 2 endpoints ----
    coeffs_cross_2_from_1 = None
    poly_order_cross = 2
    deg_min = 0.0
    deg_max = 90.0
    cross_new_col = "theta_cam_cal_xtrained"

    if calib_df is None or getattr(calib_df, "empty", True):
        print("WARNING: calib_df is empty/None -> cannot build cross-trained calibration.")
    else:
        sub1 = calib_df[(calib_df["set"] == 1) & calib_df["angle_snap_deg"].notna()].copy()
        sub2 = calib_df[(calib_df["set"] == 2) & calib_df["angle_snap_deg"].notna()].copy()

        if sub1.empty or sub2.empty:
            print("WARNING: missing set 1 or set 2 calibration rows -> cannot build cross-trained calibration.")
        elif len(sub1) < (poly_order_cross + 1):
            print("WARNING: not enough set 1 calibration points for inverse polynomial fit.")
        else:
            # ---- Fit inverse polynomial for set 1: theta -> ADC_1 ----
            theta1 = sub1["angle_snap_deg"].to_numpy(dtype=float)
            adc1   = sub1["adc_mean"].to_numpy(dtype=float)

            coeffs_inv_1 = np.polyfit(theta1, adc1, poly_order_cross)
            q1 = np.poly1d(coeffs_inv_1)

            # ---- Build dense theta grid ----
            theta_grid = np.linspace(deg_min, deg_max, 200)

            # ---- Evaluate set 1 shape in ADC space ----
            adc1_grid = q1(theta_grid)

            # ---- Determine endpoint ADCs using 0 deg and 90 deg from set 1 / set 2 ----
            # Prefer exact snapped endpoints when present; otherwise use polynomial/grid endpoints.
            sub1_0 = sub1[np.isclose(sub1["angle_snap_deg"], deg_min)]
            sub1_90 = sub1[np.isclose(sub1["angle_snap_deg"], deg_max)]
            sub2_0 = sub2[np.isclose(sub2["angle_snap_deg"], deg_min)]
            sub2_90 = sub2[np.isclose(sub2["angle_snap_deg"], deg_max)]

            if len(sub1_0) > 0:
                adc1_at_0 = float(sub1_0["adc_mean"].mean())
            else:
                adc1_at_0 = float(q1(deg_min))

            if len(sub1_90) > 0:
                adc1_at_90 = float(sub1_90["adc_mean"].mean())
            else:
                adc1_at_90 = float(q1(deg_max))

            # For set 2 endpoints, use measured means if available; otherwise fit inverse on set 2 if possible
            if len(sub2_0) > 0:
                adc2_at_0 = float(sub2_0["adc_mean"].mean())
            else:
                if len(sub2) >= (poly_order_cross + 1):
                    coeffs_inv_2 = np.polyfit(
                        sub2["angle_snap_deg"].to_numpy(dtype=float),
                        sub2["adc_mean"].to_numpy(dtype=float),
                        poly_order_cross
                    )
                    q2 = np.poly1d(coeffs_inv_2)
                    adc2_at_0 = float(q2(deg_min))
                else:
                    adc2_at_0 = float(sub2["adc_mean"].max())

            if len(sub2_90) > 0:
                adc2_at_90 = float(sub2_90["adc_mean"].mean())
            else:
                if len(sub2) >= (poly_order_cross + 1):
                    coeffs_inv_2 = np.polyfit(
                        sub2["angle_snap_deg"].to_numpy(dtype=float),
                        sub2["adc_mean"].to_numpy(dtype=float),
                        poly_order_cross
                    )
                    q2 = np.poly1d(coeffs_inv_2)
                    adc2_at_90 = float(q2(deg_max))
                else:
                    adc2_at_90 = float(sub2["adc_mean"].min())

            # ---- Normalize set 1 shape using set 1 endpoints ----
            denom1 = adc1_at_0 - adc1_at_90
            if np.isclose(denom1, 0):
                print("WARNING: set 1 endpoint span is ~0 -> cannot build cross-trained calibration.")
            else:
                adc1_norm_grid = (adc1_grid - adc1_at_90) / denom1

                # ---- Rescale normalized set 1 shape to set 2 endpoint range ----
                adc2_cross_grid = adc2_at_90 + adc1_norm_grid * (adc2_at_0 - adc2_at_90)

                # ---- Fit final cross polynomial: ADC -> theta ----
                coeffs_cross_2_from_1 = np.polyfit(adc2_cross_grid, theta_grid, poly_order_cross)

    # ---- 4) Apply cross-trained coeffs to second-application trials ----
    adc_second_theta_cross = []
    for df in adc_second_theta:
        if df is None or not hasattr(df, "columns"):
            adc_second_theta_cross.append(df)
            continue

        df2 = df.copy()

        if coeffs_cross_2_from_1 is not None and "adc_ch3" in df2.columns:
            adc_vals = pd.to_numeric(df2["adc_ch3"], errors="coerce").to_numpy(dtype=float)
            p_cross = np.poly1d(coeffs_cross_2_from_1)
            theta_cross = p_cross(adc_vals)
            theta_cross = np.clip(theta_cross, deg_min, deg_max)
            df2[cross_new_col] = theta_cross

        adc_second_theta_cross.append(df2)

    # ---- 5) Convert timestamp strings HHMMSSffffff to elapsed seconds ----
    def make_elapsed_time(df, timestamp_col="timestamp"):
        """
        Convert HHMMSSffffff timestamp strings to elapsed seconds.
        Keeps ALL rows; does not collapse repeated timestamps.
        """
        if timestamp_col not in df.columns or len(df) == 0:
            return np.arange(len(df), dtype=float)

        ts = (
            df[timestamp_col]
            .astype(str)
            .str.replace(r"\.0$", "", regex=True)
            .str.strip()
            .str.zfill(12)
        )

        ts_dt = pd.to_datetime(ts, format="%H%M%S%f", errors="coerce")

        if ts_dt.isna().all():
            return np.arange(len(df), dtype=float)

        first_valid_idx = ts_dt.first_valid_index()
        if first_valid_idx is None:
            return np.arange(len(df), dtype=float)

        t0 = ts_dt.loc[first_valid_idx]
        elapsed = (ts_dt - t0).dt.total_seconds()
        return elapsed.to_numpy(dtype=float)

    # ---- 6) Helper: get theta series for plotting ----
    def get_theta_series(
        trial_df,
        *,
        theta_col=None,
        set_id=None,
        adc_column="adc_ch3",
        clamp=True,
        deg_min=0.0,
        deg_max=90.0,
    ):
        """
        Priority:
          1) use theta_col if provided and present
          2) otherwise use self-calibration coeffs via set_id
        """
        if trial_df is None or not hasattr(trial_df, "columns"):
            return None

        df = trial_df.copy()

        # ---- y values ----
        if theta_col is not None and theta_col in df.columns:
            y = pd.to_numeric(df[theta_col], errors="coerce").to_numpy(dtype=float)
        else:
            if set_id is None or set_id not in coeffs:
                return None
            if adc_column not in df.columns:
                return None

            adc_vals = pd.to_numeric(df[adc_column], errors="coerce").to_numpy(dtype=float)
            p = np.poly1d(coeffs[set_id])
            y = p(adc_vals)

            if clamp:
                y = np.clip(y, deg_min, deg_max)

        # ---- x values ----
        if "timestamp" in df.columns:
            x = make_elapsed_time(df)
        else:
            x = np.arange(len(y), dtype=float)

        # ---- finite mask ----
        mask = np.isfinite(x) & np.isfinite(y)
        x = x[mask]
        y = y[mask]

        if len(x) == 0:
            return None

        elapsed_sec = float(x[-1] - x[0]) if len(x) > 1 else np.nan
        return x, y, elapsed_sec

    # ---- 7) Plot 3 columns ----
    n = min(3, len(adc_first_theta), len(adc_second_theta), len(adc_second_theta_cross))
    if n == 0:
        print("Nothing to plot: n=0 (your trial lists are empty).")
    else:
        fig, axes = plt.subplots(3, 3, figsize=(16, 10), sharey=True)

        for r in range(3):
            for c in range(3):
                if r >= n:
                    axes[r, c].axis("off")

        missing = 0

        for i in range(n):
            # ---- COL 1: 1st app using set 1 calibration ----
            ax = axes[i, 0]

            xy = get_theta_series(
                adc_first_theta[i],
                set_id=1,
                adc_column="adc_ch3",
                deg_min=deg_min,
                deg_max=deg_max,
            )

            if xy is None:
                missing += 1
                ax.text(0.5, 0.5, "No theta\n(1st app, set 1)", ha="center", va="center")
            else:
                x, y, elapsed_sec = xy
                ax.scatter(x, y, s=6, alpha=0.8, label=f"elapsed = {elapsed_sec:.2f} s")
                ax.legend(loc="best", fontsize=8)

            ax.set_title(f"Trial {i+1} — 1st app\nself-cal ({folder_suffix_first})")
            ax.set_ylabel("Angle (deg)")
            ax.set_xlabel("Time (s)")
            ax.set_ylim(0, 90)

            # ---- COL 2: 2nd app using set 2 calibration ----
            ax = axes[i, 1]

            xy = get_theta_series(
                adc_second_theta[i],
                set_id=2,
                adc_column="adc_ch3",
                deg_min=deg_min,
                deg_max=deg_max,
            )

            if xy is None:
                missing += 1
                ax.text(0.5, 0.5, "No theta\n(2nd app, set 2)", ha="center", va="center")
            else:
                x, y, elapsed_sec = xy
                ax.scatter(x, y, s=6, alpha=0.8, label=f"elapsed = {elapsed_sec:.2f} s")
                ax.legend(loc="best", fontsize=8)

            ax.set_title(f"Trial {i+1} — 2nd app\nself-cal ({folder_suffix_second})")
            ax.set_ylabel("Angle (deg)")
            ax.set_xlabel("Time (s)")
            ax.set_ylim(0, 90)

            # ---- COL 3: 2nd app using cross-trained set 1 -> rescaled to set 2 ----
            ax = axes[i, 2]

            xy = get_theta_series(
                adc_second_theta_cross[i],
                theta_col=cross_new_col,
                adc_column="adc_ch3",
                deg_min=deg_min,
                deg_max=deg_max,
            )

            if xy is None:
                missing += 1
                ax.text(0.5, 0.5, "No theta\n(2nd app, xtrained)", ha="center", va="center")
            else:
                x, y, elapsed_sec = xy
                ax.scatter(x, y, s=6, alpha=0.8, label=f"elapsed = {elapsed_sec:.2f} s")
                ax.legend(loc="best", fontsize=8)

            ax.set_title(f"Trial {i+1} — 2nd app\nxtrained 1→2 ({folder_suffix_second})")
            ax.set_ylabel("Angle (deg)")
            ax.set_xlabel("Time (s)")
            ax.set_ylim(0, 20)

        if missing > 0:
            print(f"NOTE: {missing} subplot(s) missing theta.")

        fig.suptitle(
            "ADAB: θ per trial\n"
            "Col 1 = 1st app self-cal | Col 2 = 2nd app self-cal | Col 3 = 2nd app using set 1 shape rescaled to set 2 endpoints",
            y=0.98
        )
        plt.tight_layout()
        plt.show()

    # ---- 8) Optional outputs you can reuse later ----
    out_cross = {
        "coeffs_cross_2_from_1": coeffs_cross_2_from_1,
        "adc_trials_second_theta_cross": adc_second_theta_cross,
    }
```

## New reapplied version
Blue Boxes
→ Self-calibrated condition

1st application trials are converted to angle using the 1st calibration curve (their own quadratic fit).

2nd application trials are converted to angle using the 2nd calibration curve (their own quadratic fit).

Red Boxes
→ Cross-trained / recalibrated condition

We take the 1st calibration curve, convert it to a normalized 0–1 scale using its own min/max ADC.

Then we rescale that normalized calibration curve to the min and max ADC of the 2nd calibration data, producing a second-application-mapped version of the first calibration curve.

This re-mapped calibration curve is used to convert 2nd application trials to angle.

# Below method could be written as a @staticmethod in analysis_adc_cam.py

```python
from typing import Optional, Sequence
from analysis_adc_cam import ADC_CAM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def run_adc_cam_pipeline_for_participant_speed(
    root_dir: str,
    speed_tag: str,
    *,
    path_to_repo: str,
    adc_column: str = "adc_ch3",
    calib_kwargs: Optional[dict] = None,
    time_unit: str = "ns",
    align_tolerance: str = "50ms",
    max_lag_samples: int = 100,  # kept in signature but NOT used for refinement below
    trial_indices_first: Optional[Sequence[int]] = None,
    trial_indices_second: Optional[Sequence[int]] = None,
    min_dlc_likelihood: float = 0.9,
    # NEW: allow specifying canonical calibration angles (deg)
    canonical_angles: Optional[Sequence[float]] = None,
) -> dict:
    """
    Run the full ADC_CAM pipeline for a given participant/date (root_dir)
    and one speed condition (slow, medm, fast, vfas, afap).

    Now also computes a cross-trained calibration for the SECOND application,
    using the first application's calibration shape (blue = self, red = cross).

    Parameters
    ----------
    ...
    min_dlc_likelihood : float, default 0.9
        DLC likelihood threshold for MCP/PIP/hand filtering.
    canonical_angles : sequence of float or None, default None
        Canonical calibration angles (deg) to use inside
        extract_calib_means_by_set. If None, that method's default
        is used (e.g. (0, 22, 45, 67, 90)).
    """

    # 1) Build ADC_CAM for this speed
    folder_suffix_first = f"B1_{speed_tag}"
    folder_suffix_second = f"B2_{speed_tag}"

    cam = ADC_CAM(
        root_dir=root_dir,
        path_to_repo=path_to_repo,
        folder_suffix_first=folder_suffix_first,
        folder_suffix_second=folder_suffix_second,
    )

    # 2) Load raw trials (paths)
    first_trials = cam.load_first()
    second_trials = cam.load_second()

    if len(first_trials) == 0 and len(second_trials) == 0:
        return {
            "abs_err_first": np.array([]),
            "abs_err_second": np.array([]),
            "abs_err_second_cross": np.array([]),
            "summary_first": pd.DataFrame(),
            "summary_second": pd.DataFrame(),
            "summary_second_cross": pd.DataFrame(),  # fixed typo
        }

    # 3) Load ADC CSVs per trial
    adc_first = cam.extract_adc_dfs_by_trial(first_trials)
    adc_second = cam.extract_adc_dfs_by_trial(second_trials)

    # --- NEW: inject canonical_angles into calib_kwargs if provided ---
    if calib_kwargs is None:
        calib_kwargs = {}
    else:
        calib_kwargs = dict(calib_kwargs)  # shallow copy so we don't mutate caller

    if canonical_angles is not None:
        calib_kwargs["canonical_angles"] = tuple(canonical_angles)

    # 4) Camera-based calibration (ADC -> theta), SELF for both sets
    calib_result = cam.calibrate_trials_with_camera(
        adc_trials_first=adc_first,
        adc_trials_second=adc_second,
        adc_column=adc_column,
        poly_order=2,
        calib_kwargs=calib_kwargs,
        clamp_theta=True,
        deg_min=0.0,
        deg_max=90.0,
    )

    calib_df         = calib_result["calib_df"]
    adc_first_theta  = calib_result["adc_trials_first_theta"]
    adc_second_theta = calib_result["adc_trials_second_theta"]

    # 4b) CROSS-trained calibration for second application
    cross_result = cam.reapply_first_calibration_to_second(
        calib_df=calib_df,
        adc_trials_second=adc_second,
        adc_column=adc_column,
        poly_order=2,
        deg_min=0.0,
        deg_max=90.0,
        new_col="theta_cam_cal_xtrained",
        clamp_theta=True,
    )
    adc_second_theta_cross = cross_result["adc_trials_second_theta_cross"]

    # 5) Trigger-time CSV per trial (timestamps)  **REPLACES FLIR MAT**
    cam_trials_first = cam.extract_trigger_time_dfs_by_trial(
        first_trials,
        add_labels=True,
        trial_base=1,
        set_label="first_cam",
        include_path=True,
    )
    cam_trials_second = cam.extract_trigger_time_dfs_by_trial(
        second_trials,
        add_labels=True,
        trial_base=1,
        set_label="second_cam",
        include_path=True,
    )

    # 6) DLC per trial
    dlc3d_trials_first = cam.extract_dlc3d_dfs_by_trial(
        first_trials,
        add_labels=True,
        trial_base=1,
        set_label="first_cam",
        include_path=True,
    )
    dlc3d_trials_second = cam.extract_dlc3d_dfs_by_trial(
        second_trials,
        add_labels=True,
        trial_base=1,
        set_label="second_cam",
        include_path=True,
    )

    # 7) Coerce DLC -> MultiIndex and compute angles
    dlc3d_trials_first  = [cam._coerce_dlc3d_multiindex(d)  for d in dlc3d_trials_first]
    dlc3d_trials_second = [cam._coerce_dlc3d_multiindex(d) for d in dlc3d_trials_second]

    aug_first,  _ = cam.compute_dlc3d_angles_by_trial(
        dlc3d_trials_first,  set_label="first_cam",  signed_in_plane=True
    )
    aug_second, _ = cam.compute_dlc3d_angles_by_trial(
        dlc3d_trials_second, set_label="second_cam", signed_in_plane=True
    )

    # 8) Attach trigger timestamps to DLC angle trials
    dlc3D_angles_first = cam.attach_cam_timestamps_to_angles(
        cam_trials_first,
        aug_first,
        time_col_name="timestamp",
        new_col_name="cam_timestamp",
    )
    dlc3D_angles_second = cam.attach_cam_timestamps_to_angles(
        cam_trials_second,
        aug_second,
        time_col_name="timestamp",
        new_col_name="cam_timestamp",
    )

    # 8b) Filter out low-likelihood DLC rows before alignment
    dlc3D_angles_first = cam.filter_angle_trials_by_likelihood(
        dlc3D_angles_first,
        bodyparts=("MCP", "PIP", "hand"),
        min_likelihood=min_dlc_likelihood,
    )
    dlc3D_angles_second = cam.filter_angle_trials_by_likelihood(
        dlc3D_angles_second,
        bodyparts=("MCP", "PIP", "hand"),
        min_likelihood=min_dlc_likelihood,
    )

    # 9) Align ADC-theta to DLC angles (time-based): SELF calibrations
    merged_first = cam.align_adc_theta_to_dlc_angles_for_set(
        dlc_angle_trials=dlc3D_angles_first,
        adc_theta_trials=adc_first_theta,
        dlc_time_col="cam_timestamp",
        adc_time_col="timestamp",
        adc_cols=["theta_cam_cal"],
        time_unit=time_unit,
        tolerance=align_tolerance,
        direction="nearest",
        suffix="_adc",
        keep_time_delta=True,
        drop_unmatched=True,
    )

    merged_second = cam.align_adc_theta_to_dlc_angles_for_set(
        dlc_angle_trials=dlc3D_angles_second,
        adc_theta_trials=adc_second_theta,
        dlc_time_col="cam_timestamp",
        adc_time_col="timestamp",
        adc_cols=["theta_cam_cal"],
        time_unit=time_unit,
        tolerance=align_tolerance,
        direction="nearest",
        suffix="_adc",
        keep_time_delta=True,
        drop_unmatched=True,
    )

    # 9b) Align ADC-theta to DLC angles (time-based): CROSS-trained calibrations
    merged_second_cross = cam.align_adc_theta_to_dlc_angles_for_set(
        dlc_angle_trials=dlc3D_angles_second,
        adc_theta_trials=adc_second_theta_cross,
        dlc_time_col="cam_timestamp",
        adc_time_col="timestamp",
        adc_cols=["theta_cam_cal_xtrained"],
        time_unit=time_unit,
        tolerance=align_tolerance,
        direction="nearest",
        suffix="_adc",
        keep_time_delta=True,
        drop_unmatched=True,
    )

    # 10) RMSE-based refinement (SELF) -- DISABLED (no further alignment)
    refined_first, summary_first = cam.refine_alignment_by_rmse_for_set(
        merged_trials=merged_first,
        dlc_col=("metric", "mcp_bend_deg", "deg"),
        adc_col="theta_cam_cal_adc",
        max_lag_samples=0,  # <- NO lag refinement
        time_col=("_t_cam_td", "", ""),
        plot_indices=None,
        set_name=f"{speed_tag.upper()} FIRST",
        do_refine=False,                 # <--- NO further alignment
    )

    refined_second, summary_second = cam.refine_alignment_by_rmse_for_set(
        merged_trials=merged_second,
        dlc_col=("metric", "mcp_bend_deg", "deg"),
        adc_col="theta_cam_cal_adc",
        max_lag_samples=0,  # <- NO lag refinement
        time_col=("_t_cam_td", "", ""),
        plot_indices=None,
        set_name=f"{speed_tag.upper()} SECOND (self)",
        do_refine=False,                 # <--- NO further alignment
    )

    # 10b) RMSE-based refinement (CROSS) -- DISABLED (no further alignment)
    refined_second_cross, summary_second_cross = cam.refine_alignment_by_rmse_for_set(
        merged_trials=merged_second_cross,
        dlc_col=("metric", "mcp_bend_deg", "deg"),
        adc_col="theta_cam_cal_xtrained_adc",
        max_lag_samples=0,  # <- NO lag refinement
        time_col=("_t_cam_td", "", ""),
        plot_indices=None,
        set_name=f"{speed_tag.upper()} SECOND (cross)",
    )

    # 11) Collect abs-error distributions using explicit trial indices
    abs_err_first = cam.collect_abs_error_for_set(
        refined_first,
        summary_first,
        dlc_col=("metric", "mcp_bend_deg", "deg"),
        adc_rmse_col="theta_cam_cal_adc_rmse",
        trial_indices=trial_indices_first,
        n_best=2,
    )

    abs_err_second = cam.collect_abs_error_for_set(
        refined_second,
        summary_second,
        dlc_col=("metric", "mcp_bend_deg", "deg"),
        adc_rmse_col="theta_cam_cal_adc_rmse",
        trial_indices=trial_indices_second,
        n_best=2,
    )

    abs_err_second_cross = cam.collect_abs_error_for_set(
        refined_second_cross,
        summary_second_cross,
        dlc_col=("metric", "mcp_bend_deg", "deg"),
        adc_rmse_col="theta_cam_cal_xtrained_adc_rmse",
        trial_indices=trial_indices_second,
        n_best=2,
    )

    return {
        "refined_first": refined_first,
        "refined_second": refined_second,
        "refined_second_cross": refined_second_cross,

        "abs_err_first": abs_err_first,
        "abs_err_second": abs_err_second,
        "abs_err_second_cross": abs_err_second_cross,
        "summary_first": summary_first,
        "summary_second": summary_second,
        "summary_second_cross": summary_second_cross,
    } 
```

```python
ERROR_THRESH_DEG = 5.0

# --- Build per-speed means (same logic as your bar plots) ---
rows = []
for i, (speed, speed_title) in enumerate(zip(speed_tags, speed_titles)):

    # 1st app (self)
    all_first = []
    # 2nd app (self)
    all_second_self  = []
    # 2nd app (cross)
    all_second_cross = []

    for p in participants:
        key = (p["name"], speed)

        a1  = results.get(key, {}).get("abs_err_first", np.array([]))
        a2s = results.get(key, {}).get("abs_err_second", np.array([]))
        a2x = results.get(key, {}).get("abs_err_second_cross", np.array([]))

        if a1.size > 0:  all_first.append(a1)
        if a2s.size > 0: all_second_self.append(a2s)
        if a2x.size > 0: all_second_cross.append(a2x)

    def _concat_finite(list_of_arrays):
        """Concatenate arrays and keep only finite values."""
        if not list_of_arrays:
            return np.array([], dtype=float)
        arr = np.concatenate(list_of_arrays).astype(float)
        return arr[np.isfinite(arr)]

    def _mae_and_n(list_of_arrays):
        """Mean absolute error + N finite."""
        arr = _concat_finite(list_of_arrays)
        if arr.size == 0:
            return np.nan, 0
        return float(np.mean(arr)), int(arr.size)

    def _pct_le_thresh(list_of_arrays, thresh=ERROR_THRESH_DEG):
        """Percent of absolute errors <= thresh + N finite."""
        arr = _concat_finite(list_of_arrays)
        if arr.size == 0:
            return np.nan, 0
        return float(100.0 * np.mean(arr <= thresh)), int(arr.size)

    # --- per-speed MAEs (you already do this) ---
    m1,  n1  = _mae_and_n(all_first)
    m2s, n2s = _mae_and_n(all_second_self)
    m2x, n2x = _mae_and_n(all_second_cross)

    # --- NEW: per-speed % <= 5° (requested) ---
    p1,  pn1  = _pct_le_thresh(all_first)
    p2x, pn2x = _pct_le_thresh(all_second_cross)

    rows.append({
        "speed": speed_title,

        "MAE 1st (self) [deg]": m1,
        "N 1st": n1,

        "MAE 2nd (self) [deg]": m2s,
        "N 2nd self": n2s,

        "MAE 2nd (cross) [deg]": m2x,
        "N 2nd cross": n2x,

        # NEW columns (only for the two sets you asked for)
        f"% ≤ {ERROR_THRESH_DEG:.1f}° 1st (self)": p1,
        f"N for % 1st": pn1,
        f"% ≤ {ERROR_THRESH_DEG:.1f}° 2nd (cross)": p2x,
        f"N for % 2nd cross": pn2x,
    })

df_speed = pd.DataFrame(rows)

# --- Average of the averages (across speeds) ---
# (mean of per-speed MAEs)
avg_1st_self  = float(df_speed["MAE 1st (self) [deg]"].mean(skipna=True))
avg_2nd_cross = float(df_speed["MAE 2nd (cross) [deg]"].mean(skipna=True))

# NEW: mean of per-speed percentages
avgpct_1st_self  = float(df_speed[f"% ≤ {ERROR_THRESH_DEG:.1f}° 1st (self)"].mean(skipna=True))
avgpct_2nd_cross = float(df_speed[f"% ≤ {ERROR_THRESH_DEG:.1f}° 2nd (cross)"].mean(skipna=True))

# --- NEW: pooled across ALL speeds (optional but usually useful) ---
# pool all finite errors across speeds, then compute pooled MAE and pooled %<=5
all_first_pooled = []
all_second_cross_pooled = []

for speed in speed_tags:
    for p in participants:
        key = (p["name"], speed)
        a1  = results.get(key, {}).get("abs_err_first", np.array([]))
        a2x = results.get(key, {}).get("abs_err_second_cross", np.array([]))
        if a1.size > 0:  all_first_pooled.append(a1)
        if a2x.size > 0: all_second_cross_pooled.append(a2x)

def _pool_stats(list_of_arrays, thresh=ERROR_THRESH_DEG):
    arr = np.concatenate(list_of_arrays).astype(float) if list_of_arrays else np.array([], dtype=float)
    arr = arr[np.isfinite(arr)]
    if arr.size == 0:
        return np.nan, np.nan, 0
    mae = float(np.mean(arr))
    pct = float(100.0 * np.mean(arr <= thresh))
    return mae, pct, int(arr.size)

pooled_mae_1st, pooled_pct_1st, pooled_n_1st = _pool_stats(all_first_pooled)
pooled_mae_2x,  pooled_pct_2x,  pooled_n_2x  = _pool_stats(all_second_cross_pooled)

df_summary = pd.DataFrame([
    {"condition": "1st application (self)",
     "avg of per-speed MAEs [deg]": avg_1st_self,
     f"avg of per-speed % ≤ {ERROR_THRESH_DEG:.1f}°": avgpct_1st_self,
     "pooled MAE [deg]": pooled_mae_1st,
     f"pooled % ≤ {ERROR_THRESH_DEG:.1f}°": pooled_pct_1st,
     "N pooled": pooled_n_1st},

    {"condition": "2nd application (cross)",
     "avg of per-speed MAEs [deg]": avg_2nd_cross,
     f"avg of per-speed % ≤ {ERROR_THRESH_DEG:.1f}°": avgpct_2nd_cross,
     "pooled MAE [deg]": pooled_mae_2x,
     f"pooled % ≤ {ERROR_THRESH_DEG:.1f}°": pooled_pct_2x,
     "N pooled": pooled_n_2x},
])

# Pretty print
pd.set_option("display.precision", 3)
display(df_speed)
display(df_summary)

print(f"Avg of per-speed MAEs — 1st (self):   {avg_1st_self:.3f} deg")
print(f"Avg of per-speed MAEs — 2nd (cross):  {avg_2nd_cross:.3f} deg")
print(f"Avg of per-speed %≤{ERROR_THRESH_DEG:.1f}° — 1st (self):  {avgpct_1st_self:.2f}%")
print(f"Avg of per-speed %≤{ERROR_THRESH_DEG:.1f}° — 2nd (cross): {avgpct_2nd_cross:.2f}%")

print(f"Pooled — 1st (self):  MAE={pooled_mae_1st:.3f} deg, %≤{ERROR_THRESH_DEG:.1f}°={pooled_pct_1st:.2f}%, N={pooled_n_1st}")
print(f"Pooled — 2nd (cross): MAE={pooled_mae_2x:.3f} deg, %≤{ERROR_THRESH_DEG:.1f}°={pooled_pct_2x:.2f}%, N={pooled_n_2x}")
```

```python
# -----------------------
# ONE number per application (across all speeds + participants)
# and ONE number across applications
# -----------------------

def _pool_all_errors(results, participants, speed_tags, key_name):
    """
    key_name: 'abs_err_first', 'abs_err_second', or 'abs_err_second_cross'
    Returns pooled finite abs errors as a 1D float array.
    """
    pooled = []
    for speed in speed_tags:
        for p in participants:
            key = (p["name"], speed)
            arr = results.get(key, {}).get(key_name, np.array([]))
            if np.size(arr) > 0:
                pooled.append(np.asarray(arr, dtype=float))
    if not pooled:
        return np.array([], dtype=float)
    pooled = np.concatenate(pooled).astype(float)
    pooled = pooled[np.isfinite(pooled)]
    return pooled

def _mae_rmse(arr):
    if arr.size == 0:
        return np.nan, np.nan, 0
    mae  = float(np.mean(arr))
    rmse = float(np.sqrt(np.mean(np.square(arr))))
    return mae, rmse, int(arr.size)

# --- POOLED (recommended): weights by number of samples ---
pooled_first = _pool_all_errors(results, participants, speed_tags, "abs_err_first")
pooled_self2 = _pool_all_errors(results, participants, speed_tags, "abs_err_second")
pooled_x2    = _pool_all_errors(results, participants, speed_tags, "abs_err_second_cross")

mae_first, rmse_first_all, n_first_all = _mae_rmse(pooled_first)
mae_self2, rmse_self2_all, n_self2_all = _mae_rmse(pooled_self2)
mae_x2,    rmse_x2_all,    n_x2_all    = _mae_rmse(pooled_x2)

print("\nPOOLED across speeds+participants:")
print(f"  1st app (self):   MAE={mae_first:.3f}°, RMSE={rmse_first_all:.3f}°, N={n_first_all}")
print(f"  2nd app (self):   MAE={mae_self2:.3f}°, RMSE={rmse_self2_all:.3f}°, N={n_self2_all}")
print(f"  2nd app (xtrain): MAE={mae_x2:.3f}°, RMSE={rmse_x2_all:.3f}°, N={n_x2_all}")

# --- Optional: "equal-weight across speeds" version (mean of per-speed values) ---
# These ignore per-speed N differences.
mean_of_speed_mae_first = float(np.nanmean(mean_first))
mean_of_speed_rmse_first = float(np.nanmean(rmse_first))

mean_of_speed_mae_self2 = float(np.nanmean(mean_second_self))
mean_of_speed_rmse_self2 = float(np.nanmean(rmse_second_self))

mean_of_speed_mae_x2 = float(np.nanmean(mean_second_cross))
mean_of_speed_rmse_x2 = float(np.nanmean(rmse_second_cross))

print("\nMEAN-OF-SPEEDS (each speed equally weighted):")
print(f"  1st app (self):   MAE={mean_of_speed_mae_first:.3f}°, RMSE={mean_of_speed_rmse_first:.3f}°")
print(f"  2nd app (self):   MAE={mean_of_speed_mae_self2:.3f}°, RMSE={mean_of_speed_rmse_self2:.3f}°")
print(f"  2nd app (xtrain): MAE={mean_of_speed_mae_x2:.3f}°, RMSE={mean_of_speed_rmse_x2:.3f}°")

# --- ONE number across applications ---
# You need to decide what "across applications" means.

# (A) Combine 1st app (self) + 2nd app (self) into one pooled "self across applications"
pooled_self_both = np.concatenate([pooled_first, pooled_self2]) if (pooled_first.size + pooled_self2.size) else np.array([], float)
mae_self_both, rmse_self_both, n_self_both = _mae_rmse(pooled_self_both)
print("\nAcross applications (SELF only: 1st+self + 2nd+self pooled):")
print(f"  SELF both apps:   MAE={mae_self_both:.3f}°, RMSE={rmse_self_both:.3f}°, N={n_self_both}")

# (B) Combine 1st app (self) + 2nd app (xtrain) into one pooled "overall across applications"
pooled_overall = np.concatenate([pooled_first, pooled_x2]) if (pooled_first.size + pooled_x2.size) else np.array([], float)
mae_overall, rmse_overall, n_overall = _mae_rmse(pooled_overall)
print("\nAcross applications (1st self + 2nd xtrain pooled):")
print(f"  OVERALL:          MAE={mae_overall:.3f}°, RMSE={rmse_overall:.3f}°, N={n_overall}")

# (C) If you literally want to average the two single numbers (equal-weight apps),
# use the mean of the two pooled metrics (not pooling raw errors):
mae_avg_apps = float(np.nanmean([mae_first, mae_self2]))
rmse_avg_apps = float(np.nanmean([rmse_first_all, rmse_self2_all]))
print("\nEqual-weight average of application-level pooled metrics (1st self & 2nd self):")
print(f"  avg apps:         MAE={mae_avg_apps:.3f}°, RMSE={rmse_avg_apps:.3f}°")
```

```python
# -------------------------------------------------
# Threshold sweep: % of errors ≤ θ (θ = 1° … 15°)
# Combined across 1st (self) + 2nd (cross)
# -------------------------------------------------

THRESHOLDS = np.arange(1.0, 16.0, 1.0)  # 1°, 2°, ..., 15°

# -------------------------------------------------
# Collect pooled errors across ALL speeds + participants
# -------------------------------------------------
all_first_pooled = []
all_second_cross_pooled = []

for speed in speed_tags:
    for p in participants:
        key = (p["name"], speed)

        a1  = results.get(key, {}).get("abs_err_first", np.array([]))
        a2x = results.get(key, {}).get("abs_err_second_cross", np.array([]))

        if np.size(a1) > 0:
            all_first_pooled.append(np.asarray(a1, dtype=float))
        if np.size(a2x) > 0:
            all_second_cross_pooled.append(np.asarray(a2x, dtype=float))

def _finite_concat(list_of_arrays):
    if not list_of_arrays:
        return np.array([], dtype=float)
    arr = np.concatenate(list_of_arrays).astype(float)
    return arr[np.isfinite(arr)]

def _pct_le(arr, thr):
    arr = np.asarray(arr, dtype=float)
    arr = arr[np.isfinite(arr)]
    return np.nan if arr.size == 0 else float(100.0 * np.mean(arr <= thr))

# -------------------------------------------------
# Pooled arrays
# -------------------------------------------------
first_errs = _finite_concat(all_first_pooled)
cross_errs = _finite_concat(all_second_cross_pooled)

both_errs = (
    np.concatenate([first_errs, cross_errs])
    if (first_errs.size + cross_errs.size) > 0
    else np.array([], dtype=float)
)

print(f"N pooled 1st (self)         = {first_errs.size}")
print(f"N pooled 2nd (cross)        = {cross_errs.size}")
print(f"N pooled BOTH (1st + 2ndx)  = {both_errs.size}")

# -------------------------------------------------
# ONE pooled curve: % ≤ θ across BOTH conditions
# -------------------------------------------------
print("\nPOOLED across 1st(self) + 2nd(cross):")
rows_pooled = []

for thr in THRESHOLDS:
    pct = _pct_le(both_errs, thr)
    rows_pooled.append({
        "threshold_deg": thr,
        "pct_le": pct,
        "N": both_errs.size
    })
    print(f"  % ≤ {thr:2.0f}° : {pct:6.2f}%")

df_pooled_curve = pd.DataFrame(rows_pooled)

# -------------------------------------------------
# Per-speed threshold sweep
# -------------------------------------------------
rows_both_pct = []

for speed, speed_title in zip(speed_tags, speed_titles):

    first_list = []
    cross_list = []

    for p in participants:
        key = (p["name"], speed)
        a1  = results.get(key, {}).get("abs_err_first", np.array([]))
        a2x = results.get(key, {}).get("abs_err_second_cross", np.array([]))
        if np.size(a1) > 0:  first_list.append(np.asarray(a1, dtype=float))
        if np.size(a2x) > 0: cross_list.append(np.asarray(a2x, dtype=float))

    first_speed = _finite_concat(first_list)
    cross_speed = _finite_concat(cross_list)
    both_speed  = (
        np.concatenate([first_speed, cross_speed])
        if (first_speed.size + cross_speed.size) > 0
        else np.array([], dtype=float)
    )

    row = {"speed": speed_title}
    for thr in THRESHOLDS:
        row[f"% ≤ {thr:.0f}°"] = _pct_le(both_speed, thr)

    rows_both_pct.append(row)

df_both_pct = pd.DataFrame(rows_both_pct)

# -------------------------------------------------
# Average across speeds (one number per angle)
# -------------------------------------------------
rows_avg = []

for thr in THRESHOLDS:
    col = f"% ≤ {thr:.0f}°"
    rows_avg.append({
        "threshold_deg": thr,
        "avg_pct_across_speeds": float(df_both_pct[col].mean(skipna=True))
    })

df_avg_curve = pd.DataFrame(rows_avg)

# -------------------------------------------------
# Display tables
# -------------------------------------------------
display(df_both_pct)
display(df_avg_curve)

print("\nAVG across speeds (1st+self + 2nd+cross):")
for _, r in df_avg_curve.iterrows():
    print(f"  % ≤ {r['threshold_deg']:2.0f}° : {r['avg_pct_across_speeds']:6.2f}%")
```

```python
def min_threshold_for_target_pct(errs, target_pct=90.0, step_deg=1.0, max_deg=90.0):
    """
    Return the smallest threshold (deg) such that >= target_pct of errs are <= threshold.
    Increments by step_deg until max_deg.
    """
    arr = np.asarray(errs, dtype=float)
    arr = arr[np.isfinite(arr)]
    if arr.size == 0:
        return np.nan, np.nan, 0  # threshold, achieved_pct, N

    thr = 0.0
    while thr <= max_deg + 1e-9:
        achieved = 100.0 * np.mean(arr <= thr)
        if achieved >= target_pct:
            return thr, achieved, int(arr.size)
        thr += step_deg

    # If not reached within max_deg
    return np.nan, 100.0 * np.mean(arr <= max_deg), int(arr.size)

TARGET_PCT = 90.0

thr, achieved, N = min_threshold_for_target_pct(both_errs, target_pct=TARGET_PCT, step_deg=1.0, max_deg=90.0)

if np.isfinite(thr):
    print(f"Smallest threshold with ≥{TARGET_PCT:.0f}% of errors ≤ threshold: {thr:.0f}° "
          f"(achieved {achieved:.2f}%, N={N})")
else:
    print(f"Did not reach ≥{TARGET_PCT:.0f}% within 0–90°. Achieved {achieved:.2f}% at 90° (N={N}).")
```
