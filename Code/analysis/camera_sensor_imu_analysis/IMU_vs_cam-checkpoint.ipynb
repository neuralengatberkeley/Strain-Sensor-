```python
def pct_removed_single_pass(raw_df, angle_col, max_delta_deg, unwrap=True, verbose=True):
    """
    Single-pass jump filter:
        - compute diffs once
        - drop rows once
        - no cascading iterations
    """

    if raw_df is None or raw_df.empty:
        return None

    df = raw_df.copy()
    n_initial = len(df)

    x = df[angle_col].to_numpy(dtype=float)

    if unwrap:
        x = np.rad2deg(np.unwrap(np.deg2rad(x)))

    bad = np.zeros_like(x, dtype=bool)
    bad[1:] = np.abs(np.diff(x)) > max_delta_deg

    n_removed = int(bad.sum())
    df_clean = df.loc[~bad].reset_index(drop=True)

    pct_removed = 100.0 * n_removed / n_initial if n_initial > 0 else np.nan

    if verbose:
        print(f"SINGLE PASS removed {n_removed} / {n_initial} "
              f"({pct_removed:.2f}%)")

    return {
        "initial_rows": n_initial,
        "final_rows": len(df_clean),
        "removed": n_removed,
        "pct_removed": pct_removed,
        "clean_df": df_clean,
    }


TRIAL_LEN_SEC = 10.0

# Compute bend/pitch/azimuth ONCE
aug_imu_first, tall_first = cam.compute_imu_bend_pitch_azimuth_first(
    quat_cols=("euler1", "euler2"),
    fixed_axis="y",
    moving_axis="y",
    plane_normal_axis="z",
    quat_order="wxyz",
    bend_col="imu_bend_deg",
    pitch_col="imu_pitch_deg",
    azim_col="imu_azimuth_deg",
    time_col="t_sec",
    trial_len_sec=TRIAL_LEN_SEC,
    zero_baseline_bend=True,
    zero_baseline_pitch=True,
    zero_baseline_azim=False,
)

aug_imu_second, tall_second = cam.compute_imu_bend_pitch_azimuth_second(
    quat_cols=("euler1", "euler2"),
    fixed_axis="y",
    moving_axis="y",
    plane_normal_axis="z",
    quat_order="wxyz",
    bend_col="imu_bend_deg",
    pitch_col="imu_pitch_deg",
    azim_col="imu_azimuth_deg",
    time_col="t_sec",
    trial_len_sec=TRIAL_LEN_SEC,
    zero_baseline_bend=True,
    zero_baseline_pitch=True,
    zero_baseline_azim=False,
)

# Freeze "before jump-filter"
aug_imu_first_raw  = [df.copy() if df is not None else None for df in aug_imu_first]
aug_imu_second_raw = [df.copy() if df is not None else None for df in aug_imu_second]

BEND_COL = "imu_azimuth_deg"

trial_i = 0

print("\nFIRST trial single-pass analysis")
stats_first = pct_removed_single_pass(
    aug_imu_first_raw[trial_i],
    angle_col=BEND_COL,
    max_delta_deg=20,
    unwrap=True,
)

print("\nSECOND trial single-pass analysis")
stats_second = pct_removed_single_pass(
    aug_imu_second_raw[trial_i],
    angle_col=BEND_COL,
    max_delta_deg=20,
    unwrap=True,
)

# Apply your class jump filter (NOTE: may still be wrap-sensitive unless you add unwrap inside it)
aug_imu_first  = cam.filter_joint_angle_jumps_in_trials(
    aug_imu_first,
    angle_col=BEND_COL,
    max_delta_deg=20,
    verbose=False,
)
aug_imu_second = cam.filter_joint_angle_jumps_in_trials(
    aug_imu_second,
    angle_col=BEND_COL,
    max_delta_deg=20,
    verbose=False,
)

# Baseline zeroing
aug_imu_first = IMU_cam.zero_baseline_trials_windowed(
    aug_imu_first,
    angle_col=BEND_COL,
    n_points=40,
    use_median=True,
    abs_values=True,
)
aug_imu_second = IMU_cam.zero_baseline_trials_windowed(
    aug_imu_second,
    angle_col=BEND_COL,
    n_points=40,
    use_median=True,
    abs_values=True,
)
```

```python
# -------------------------------------------------
# IMU summary stats: MAE + RMSE
# per application, then average across applications
# -------------------------------------------------

def _finite(arr):
    arr = np.asarray(arr, dtype=float)
    return arr[np.isfinite(arr)]

def _mae_rmse(arr):
    if arr.size == 0:
        return np.nan, np.nan, 0
    mae  = float(np.mean(arr))
    rmse = float(np.sqrt(np.mean(np.square(arr))))
    return mae, rmse, int(arr.size)

# ---- Pool errors across ALL speeds + participants ----
first_all  = []
second_all = []

for speed in speed_tags:
    for p in participants:
        pname = p["name"]
        if not include_participant(pname, speed):
            continue

        key = (pname, speed)

        a1 = results_imu.get(key, {}).get("abs_err_first", np.array([]))
        a2 = results_imu.get(key, {}).get("abs_err_second", np.array([]))

        if a1.size > 0:
            first_all.append(_finite(a1))
        if a2.size > 0:
            second_all.append(_finite(a2))

first_all  = _finite(np.concatenate(first_all))  if first_all  else np.array([], float)
second_all = _finite(np.concatenate(second_all)) if second_all else np.array([], float)

# ---- Per-application MAE / RMSE ----
mae_1, rmse_1, n1 = _mae_rmse(first_all)
mae_2, rmse_2, n2 = _mae_rmse(second_all)

# ---- Average across applications (equal-weight) ----
avg_mae_apps  = float(np.nanmean([mae_1,  mae_2]))
avg_rmse_apps = float(np.nanmean([rmse_1, rmse_2]))

# ---- Print summary ----
print("\nIMU ANGULAR ERROR SUMMARY")
print("-------------------------")
print(f"FIRST application : MAE = {mae_1:.3f}°, RMSE = {rmse_1:.3f}°, N = {n1}")
print(f"SECOND application: MAE = {mae_2:.3f}°, RMSE = {rmse_2:.3f}°, N = {n2}")
print(f"AVERAGE (2 apps)  : MAE = {avg_mae_apps:.3f}°, RMSE = {avg_rmse_apps:.3f}°")
```

```python
import numpy as np
import pandas as pd

# ----------------------------------------
# CONFIG
# ----------------------------------------
SPEED_SUBSET = speed_tags[:2]        # first two speeds only
ANGLE_MAX = 40
ANGLES = np.arange(1, ANGLE_MAX + 1, 1)
PERCENTILES = [70, 90, 99]

# ----------------------------------------
# Helpers
# ----------------------------------------
def _finite(arr):
    arr = np.asarray(arr, dtype=float)
    return arr[np.isfinite(arr)]

def pooled_abs_errors_subset(results_dict, which):
    """Pool finite abs errors across participants and FIRST TWO speeds."""
    all_errs = []
    for speed in SPEED_SUBSET:
        for p in participants:
            pname = p["name"]
            if not include_participant(pname, speed):
                continue
            arr = results_dict.get((pname, speed), {}).get(which, np.array([]))
            if arr is None or np.size(arr) == 0:
                continue
            arr = _finite(arr)
            if arr.size:
                all_errs.append(arr)
    return np.concatenate(all_errs) if all_errs else np.array([], float)

def mae_rmse(arr):
    if arr.size == 0:
        return np.nan, np.nan, 0
    mae  = float(np.mean(arr))
    rmse = float(np.sqrt(np.mean(np.square(arr))))
    return mae, rmse, int(arr.size)

def pct_le(arr, thr):
    if arr.size == 0:
        return np.nan
    return float(100.0 * np.mean(arr <= thr))

# ----------------------------------------
# Pool errors (FIRST TWO speeds only)
# ----------------------------------------
errs_first  = pooled_abs_errors_subset(results_imu, "abs_err_first")
errs_second = pooled_abs_errors_subset(results_imu, "abs_err_second")
errs_both   = (
    np.concatenate([errs_first, errs_second])
    if (errs_first.size + errs_second.size) > 0
    else np.array([], float)
)

print(f"N pooled FIRST  (2 speeds) = {errs_first.size}")
print(f"N pooled SECOND (2 speeds) = {errs_second.size}")
print(f"N pooled BOTH   (2 speeds) = {errs_both.size}")

# ----------------------------------------
# MAE / RMSE summary
# ----------------------------------------
mae_1, rmse_1, n1 = mae_rmse(errs_first)
mae_2, rmse_2, n2 = mae_rmse(errs_second)
mae_b, rmse_b, nb = mae_rmse(errs_both)

print("\nMAE / RMSE (FIRST TWO speeds only)")
print("---------------------------------")
print(f"FIRST  app : MAE = {mae_1:.3f}°, RMSE = {rmse_1:.3f}°, N = {n1}")
print(f"SECOND app : MAE = {mae_2:.3f}°, RMSE = {rmse_2:.3f}°, N = {n2}")
print(f"BOTH apps  : MAE = {mae_b:.3f}°, RMSE = {rmse_b:.3f}°, N = {nb}")

# ----------------------------------------
# Percentiles (pooled)
# ----------------------------------------
print("\nPercentile thresholds (FIRST TWO speeds only):")
for p in PERCENTILES:
    thr = np.percentile(errs_both, p) if errs_both.size else np.nan
    print(f"  {p:2d}th percentile = {thr:.2f}°")

# ----------------------------------------
# Angle → percentile curve (1–40°)
# ----------------------------------------
rows = []
for ang in ANGLES:
    rows.append({
        "angle_deg": ang,
        "FIRST_pct_le":  pct_le(errs_first,  ang),
        "SECOND_pct_le": pct_le(errs_second, ang),
        "BOTH_pct_le":   pct_le(errs_both,   ang),
    })

df_angle_pct_2speeds = pd.DataFrame(rows)

display(df_angle_pct_2speeds)

print("\nAngle → % ≤ angle (FIRST TWO speeds only)")
print("----------------------------------------")
for _, r in df_angle_pct_2speeds.iterrows():
    print(
        f"{int(r.angle_deg):2d}° : "
        f"FIRST={r.FIRST_pct_le:6.2f}% | "
        f"SECOND={r.SECOND_pct_le:6.2f}% | "
        f"BOTH={r.BOTH_pct_le:6.2f}%"
    )
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

# ---------------------------
# Match your exclusion rule
# ---------------------------
EXCLUDE = {("P4", "afap")}

def include_participant(pname, speed):
    return (pname, speed) not in EXCLUDE

# ---------------------------
# Build participant-level mean MAE per speed (per application)
# ---------------------------
rows = []
speed_to_idx = {s: i for i, s in enumerate(speed_tags)}  # 0..K-1

for speed in speed_tags:
    sidx = speed_to_idx[speed]
    for p in participants:
        pname = p["name"]
        if not include_participant(pname, speed):
            continue

        res = results_imu.get((pname, speed), {})

        # First application mean
        a1 = np.asarray(res.get("abs_err_first", []), dtype=float)
        a1 = a1[np.isfinite(a1)]
        if a1.size:
            rows.append({
                "participant": pname,
                "speed": speed,
                "speed_idx": float(sidx),
                "application": "first",
                "mae_mean": float(np.mean(a1)),
                "n_trials": int(a1.size),
            })

        # Second application mean
        a2 = np.asarray(res.get("abs_err_second", []), dtype=float)
        a2 = a2[np.isfinite(a2)]
        if a2.size:
            rows.append({
                "participant": pname,
                "speed": speed,
                "speed_idx": float(sidx),
                "application": "second",
                "mae_mean": float(np.mean(a2)),
                "n_trials": int(a2.size),
            })

dfm = pd.DataFrame(rows)

print("Participant-level mean MAE rows (this is N for the mean-level mixed model):")
print(dfm.groupby("application").size())
display(dfm.sort_values(["application", "participant", "speed_idx"]).head(15))

# ---------------------------
# Fit mean-level LME per application (like your strain-gauge model)
#   mae_mean ~ speed_idx + (1|participant)
# ---------------------------
def fit_mean_level(app):
    dfa = dfm[dfm["application"] == app].copy()

    # Guard: need at least 2 participants and 2 speeds worth of rows
    if dfa["participant"].nunique() < 2 or dfa["speed_idx"].nunique() < 2:
        print(f"\nNot enough data to fit mixed model for {app}.")
        return None

    fit = smf.mixedlm("mae_mean ~ speed_idx", dfa, groups=dfa["participant"]).fit(reml=True)

    slope = float(fit.params["speed_idx"])
    zval  = float(fit.tvalues["speed_idx"])
    pval  = float(fit.pvalues["speed_idx"])
    nobs  = int(dfa.shape[0])

    print(f"\n=== {app.upper()} application (MEAN-LEVEL) ===")
    print(f"slope={slope:.3f}, z={zval:.3f}, N={nobs}, p={pval:.3g}")

    print("\nManuscript-ready text:")
    print(
        "A linear mixed-effects model with speed as a fixed effect and participant as a random intercept "
        f"revealed {'no ' if pval>=0.05 else ''}significant effect of speed on mean IMU MAE for the {app} application "
        f"(slope = {slope:.3f}, z = {zval:.3f}, N = {nobs}, p = {pval:.3g})."
    )

    return fit

m_first_mean  = fit_mean_level("first")
m_second_mean = fit_mean_level("second")

# Optional: see the full summaries
# if m_first_mean is not None:  print(m_first_mean.summary())
# if m_second_mean is not None: print(m_second_mean.summary())


# ---------------------------
# OPTIONAL: If you instead want "differences by speed" without assuming linear trend
# (categorical speed). This answers a slightly different question.
# ---------------------------
# for app in ["first", "second"]:
#     dfa = dfm[dfm["application"] == app].copy()
#     fit_cat = smf.mixedlm("mae_mean ~ C(speed)", dfa, groups=dfa["participant"]).fit(reml=True)
#     print(f"\n--- {app.upper()} categorical speed model summary ---")
#     print(fit_cat.summary())
```

```python
# ---------------------------
# Collapse across applications (keep speed!)
# One row per participant × speed
# ---------------------------

# Pivot so first/second are columns (mae_mean and n_trials)
dfp = dfm.pivot_table(
    index=["participant", "speed", "speed_idx"],
    columns="application",
    values=["mae_mean", "n_trials"],
    aggfunc="first"
)

# flatten MultiIndex columns: ('mae_mean','first') -> 'mae_mean_first'
dfp.columns = [f"{a}_{b}" for a, b in dfp.columns]
dfp = dfp.reset_index()

# --- Option A: unweighted average of the two application means (if both exist) ---
dfp["mae_mean_both_unweighted"] = dfp[["mae_mean_first", "mae_mean_second"]].mean(axis=1, skipna=True)

# --- Option B (recommended): weighted by number of trials in each application ---
w1 = dfp["n_trials_first"].fillna(0.0)
w2 = dfp["n_trials_second"].fillna(0.0)
m1 = dfp["mae_mean_first"]
m2 = dfp["mae_mean_second"]

den = (w1 + w2)
dfp["mae_mean_both_weighted"] = np.where(
    den > 0,
    (m1.fillna(0.0) * w1 + m2.fillna(0.0) * w2) / den,
    np.nan
)

# Choose which combined outcome you want to analyze:
OUTCOME_COL = "mae_mean_both_weighted"   # <-- recommended
# OUTCOME_COL = "mae_mean_both_unweighted"

df_both = dfp[np.isfinite(dfp[OUTCOME_COL])].copy()

print("Rows after collapsing across applications (N for combined model):", df_both.shape[0])
display(df_both.sort_values(["participant", "speed_idx"]).head(15))

# ---------------------------
# Fit combined mean-level LME:
#   MAE_mean_combined ~ speed_idx + (1|participant)
# ---------------------------
if df_both["participant"].nunique() < 2 or df_both["speed_idx"].nunique() < 2:
    print("Not enough data to fit combined model.")
else:
    fit_both = smf.mixedlm(
        f"{OUTCOME_COL} ~ speed_idx",
        df_both,
        groups=df_both["participant"]
    ).fit(reml=True)

    slope = float(fit_both.params["speed_idx"])
    zval  = float(fit_both.tvalues["speed_idx"])
    pval  = float(fit_both.pvalues["speed_idx"])
    nobs  = int(df_both.shape[0])

    print("\n=== BOTH applications collapsed (MEAN-LEVEL) ===")
    print(f"slope={slope:.3f}, z={zval:.3f}, N={nobs}, p={pval:.3g}")

    print("\nManuscript-ready text:")
    print(
        "After averaging MAE across the first and second applications within each participant and movement speed, "
        "a linear mixed-effects model with speed as a fixed effect and participant as a random intercept "
        f"revealed {'no ' if pval>=0.05 else ''}significant effect of speed on mean IMU MAE "
        f"(slope = {slope:.3f}, z = {zval:.3f}, N = {nobs}, p = {pval:.3g})."
    )

    # Optional: full summary
    # print(fit_both.summary())

```
