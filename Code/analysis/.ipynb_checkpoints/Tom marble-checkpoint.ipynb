```python
adc_trials_first = bb.extract_adc_dfs_by_trial(df_first)  # list of 15 dataframes (index 0..14)

adc_trials_second = bb.extract_adc_dfs_by_trial(df_second)
plt.figure(figsize=(6,4))
for df in adc_trials_first:  # blue = first
    if df.empty or "adc_ch3" not in df.columns: 
        continue
    y = pd.to_numeric(df["adc_ch3"], errors="coerce"); n = len(y)
    x = np.linspace(0,10,n); m = y.notna(); yv = y[m]
    y_norm = (yv - yv.min()) / (yv.max()-yv.min()) if (yv.max()>yv.min()) else (yv*0)
    plt.plot(x[m], y_norm, ".", markersize=2, color="blue")
for df in adc_trials_second:  # red = second
    if df.empty or "adc_ch3" not in df.columns: 
        continue
    y = pd.to_numeric(df["adc_ch3"], errors="coerce"); n = len(y)
    x = np.linspace(0,10,n); m = y.notna(); yv = y[m]
    y_norm = (yv - yv.min()) / (yv.max()-yv.min()) if (yv.max()>yv.min()) else (yv*0)
    plt.plot(x[m], y_norm, ".", markersize=2, color="red")
plt.xlabel("Time (s)"); plt.ylabel("Normalized ADC (0–1)")
plt.title("ADC (adc_ch3) — First set blue, Second set red")
plt.tight_layout(); plt.show()
```
